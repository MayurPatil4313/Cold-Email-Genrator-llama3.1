WebBaseLoader in Langchain
This covers how to use WebBaseLoader to load all text from HTML webpages into a document format that we can use downstream
https://python.langchain.com/docs/integrations/document_loaders/web_base/#load

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/lead-data-scientist/job/R-62512")
```

```python
doc = loader.load().pop().metadata
print(doc)
```

```python
doc = loader.load().pop().page_content
print(doc)
```
